# EDA Sample

### 번호 | 시각화 방법 | 비교/대상
#### 1 | 막대그래프 (Count Plot) | 품종별 샘플 개수 (Name)
#### 2 | 박스플롯 (Box Plot) | 각 수치형 변수 개별 분포
#### 3 | 이중 박스플롯 (Grouped Box Plot) | Name별 SepalLength 분포
#### 4 | 바이올린 플롯 (Violin Plot) | Name별 PetalLength 분포
#### 5 | 히스토그램+KDE (Histplot) | 각 수치형 변수 개별 분포
#### 6 | 산점도 (Scatter Plot) | SepalLength vs SepalWidth (품종별 색상)
#### 7 | 산점도 (Scatter Plot) | PetalLength vs PetalWidth (품종별 색상)
#### 8 | 히트맵 (Heatmap) | 수치형 변수 간 상관관계

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C://Ai-class/iris.csv')
```

```python
print(df.info())
print(df.describe())
```

### 1. Name(품종)별 데이터 개수 시각화 (막대그래프)

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Name', data=df)
plt.title('Number of Samples per Iris Species', fontsize=16)
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

### 2. 수치형 컬럼별 박스플롯 (SepalLength, SepalWidth, PetalLength, PetalWidth)

```python
num_cols = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']

plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols):
    plt.subplot(2, 2, i+1)
    sns.boxplot(y=df[col])
    plt.title(f'{col} Distribution (Boxplot)')
plt.tight_layout()
plt.show()
```

### 3. 이중 박스플롯 (Species별 SepalLength 분포)

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Name', y='SepalLength', data=df)
plt.title('SepalLength by Species (Boxplot)', fontsize=16)
plt.xlabel('Species')
plt.ylabel('SepalLength')
plt.show()
```

### 4. 바이올린 플롯 (Species별 PetalLength 분포)

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Name', y='PetalLength', data=df, inner='box')
plt.title('PetalLength by Species (Violin Plot)', fontsize=16)
plt.xlabel('Species')
plt.ylabel('PetalLength')
plt.show()
```

### 5. 각 수치형 변수별 히스토그램+KDE

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'{col} Histogram with KDE')
plt.tight_layout()
plt.show()
```

### 6. SepalLength vs SepalWidth 산점도 (품종별 색상)

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLength', y='SepalWidth', hue='Name', data=df, palette='viridis')
plt.title('SepalLength vs SepalWidth by Species', fontsize=16)
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend(title='Species')
plt.show()
```

### 7. PetalLength vs PetalWidth 산점도 (품종별 색상)

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PetalLength', y='PetalWidth', hue='Name', data=df, palette='coolwarm')
plt.title('PetalLength vs PetalWidth by Species', fontsize=16)
plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
plt.legend(title='Species')
plt.show()
```

### 8. 상관관계 히트맵

```python
plt.figure(figsize=(10, 8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Features', fontsize=16)
plt.show()
```

```python
# 상관관계 행렬 출력
print("\n상관계수 행렬:")
print(corr)
```
